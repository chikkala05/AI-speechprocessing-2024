```python
import librosa
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
import librosa.display   
```

```python
# A1. Loading the Audio file and Plotting the Graph of Speech Signal.
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

filename = "E:\AUDIO.wav"
y, sr = librosa.load(filename)

plt.figure(figsize=(10, 4))
librosa.display.waveshow(y, sr=sr, color='black') 

plt.title('Speech Signal Waveform')
plt.xlabel('Time (samples)')
plt.ylabel('Amplitude')
plt.show() 
```

```python
# A2. Finding the Length and Magnitude range fro the signal. 
```

```python
length_seconds = librosa.get_duration(y=y, sr=sr)
magnitude_range = max(y) - min(y)
print(f"Length of the signal: {length_seconds:.2f} seconds")
print(f"Magnitude range of the signal: {magnitude_range:.2f}")
```

```python
# A3. Segment of the speech from the signal.
```

```python
segment_start = 2
segment_duration = 5  
segment = y[int(segment_start * sr):int((segment_start + segment_duration) * sr)]
ipd.display(ipd.Audio(segment, rate=sr)) 
```

```python
# A4. Comparision of Original and Abruptly Segmented Speech signals. 
```

```python
start_time = 2
end_time = 5
y_abrupt = y.copy()
y_abrupt[int(start_time * sr):int(end_time * sr)] = 0
 
plt.figure(figsize=(10, 4))
plt.plot(y, label='Original')
plt.plot(y_abrupt, label='Abruptly Segmented')
plt.title('Comparison of Original and Abruptly Segmented Speech Signals')
plt.xlabel('Time (samples)')
plt.ylabel('Amplitude')
plt.legend()
plt.show() 
```
